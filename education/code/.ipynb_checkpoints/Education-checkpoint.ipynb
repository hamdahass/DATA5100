## Data Preparation
Select relevant subsets of the data. We will get only the columns we need from the school information datatset (year, school identity, location, and school type information).

```python
school_information = school_information[['SCHOOL_YEAR', 'NCESSCH', 'LSTATE', 'LZIP', 'SCH_TYPE_TEXT', 'LEVEL', 'CHARTER_TEXT']]
```

```python
school_information.head()
```

```python
## Rename the columns of both datasets
This allows us to follow best practices of style. 
```

```python
edgap = edgap.rename(
    columns={
        "NCESSCH School ID": "id",
        "CT Pct Adults with College Degree": "percent_college",
        "CT Unemployment Rate": "rate_unemployment",
        "CT Pct Childre In Married Couple Family": "percent_married",
        "CT Median Household Income": "median_income",
        "School ACT average (or equivalent if SAT score)": "average_act",
        "School Pct Free and Reduced Lunch": "percent_lunch",
    }
)
```

```python
school_information = school_information.rename(
    columns={
        "SCHOOL_YEAR": "year",
        "NCESSCH": "id",
        "LSTATE": "state",
        "LZIP": "zip_code",
        "SCH_TYPE_TEXT": "school_type",
        "LEVEL": "school_level",
        "CHARTER_TEXT": "charter"
    }
)
```

```python
edgap.head()
```

```python
school_information.head()
```

## Join data Frames
We will join the dataframe using the identity of the school as the key. The identity is the NCESSCH.
In the edgap data, the NCESSCH is an object and in the school data, its a float, so we have to make sure 
the types match (Cast the school data as an object).

```python
school_information['id'] = school_information['id'].astype('object')
```

```python
school_information.info()
```

## Join the data frames and call the result df.
We will do a left join, so that we keep all of the information from the edgap dataset and bring in information from the school dataset where we have information about the socioeconomic varibales and act scores.

```python
df = edgap.merge(school_information, how='left', on='id')
```

## DATA imputation
We will take the data we have and replace the nans with values that we can estimate from the data.
We will start with defining predictor variables that we want to use in our analysis to predict the average act score. We will impute the missing values of these variables.

```python
predictor_variables = [
    'rate_unemployment',
    'percent_college',
    'percent_married',
    'median_income',
    'percent_lunch',
    'state',
    'charter'
]
```

## We will use iterative imputer to replace the missing values.
Define the imputer

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
```

- Fit the imputer using the numerical predictor variables.

```python
numerical_predictors = df[predictor_variables].select_dtypes(include='number').columns.to_list()
print(numerical_predictors)
```

```python
## Fit the imputer with the data from those columns.
```

```python
imputer.fit(df.loc[:, numerical_predictors])
```

```python
## Apply transform() method
- We will apply the transform method to replace the values
```

```python
df.loc[:, numerical_predictors] = imputer.transform(df.loc[:, numerical_predictors])
```

```python
df.isna().sum().to_frame(name='Number of Missing Values')
```
